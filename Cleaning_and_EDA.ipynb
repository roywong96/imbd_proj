# Genres

```python
# Raw data to visualise genres
film = pd.read_csv("/Users/roywong/Desktop/Work_Stuff/PythonPortfolio/10.web_imdb_data/movies_genres_unchanged.csv")

#genre parsing
import ast
film.genres = film.genres.apply(lambda x : ast.literal_eval(x))


# Getting distinct genre types for generating columns of genre type.
genre_columns = list(set([j for i in film['genres'].tolist() for j in i]))
```

```python
# Using original dataset to produce more visualisation
genre_columns
```

```python
# I create a new list using for loop, to check how many times uniqe genre was repaeted 
# and then convert it to a dataframe
total= []
for genre in genre_columns:
    _=film.genres.apply(lambda x: genre in x)
    total.append([genre, _.sum()])
    #print(genre, _.sum())
total = pd.DataFrame(total,columns = ['Genre','Sum'])
total = total.sort_values(by='Sum',ascending=False)
total
```

```python
plt.title("Populiarity acording to IMDB 2000 popular movies of 2000-2020")
sns.barplot(y=total.Genre, x=total.Sum)
```

```python
# tracking genres by year
df_cat_year = df[['Comedy', 'Action', 'Thriller', 'Fantasy', 'Drama', 
                         'Western', 'Biography', 'Mystery', 'Musical', 'War', 
                         'Sci-Fi', 'Sport', 'Music', 'Horror', 'Crime', 'Adventure', 
                         'Family', 'Animation', 'History', 'Romance', 'Year']]


# Created 20 new dataframes for individual genre to check how many times they appear every year
action = df_cat_year.groupby(['Year']).Action.agg([sum])
drama = df_cat_year.groupby(['Year']).Drama.agg([sum])
horror = df_cat_year.groupby(['Year']).Horror.agg([sum])
thriller = df_cat_year.groupby(['Year']).Thriller.agg([sum])
comedy = df_cat_year.groupby(['Year']).Comedy.agg([sum])
romance = df_cat_year.groupby(['Year']).Romance.agg([sum])
adventure = df_cat_year.groupby(['Year']).Adventure.agg([sum])
crime = df_cat_year.groupby(['Year']).Crime.agg([sum])
fantasy = df_cat_year.groupby(['Year']).Fantasy.agg([sum])
western = df_cat_year.groupby(['Year']).Western.agg([sum])
biography = df_cat_year.groupby(['Year']).Biography.agg([sum])
mystery = df_cat_year.groupby(['Year']).Mystery.agg([sum])
musical = df_cat_year.groupby(['Year']).Musical.agg([sum])
war = df_cat_year.groupby(['Year']).War.agg([sum])
sci_fi = df_cat_year.groupby(['Year'])['Sci-Fi'].agg([sum])
sport = df_cat_year.groupby(['Year']).Sport.agg([sum])
music = df_cat_year.groupby(['Year']).Music.agg([sum])
family = df_cat_year.groupby(['Year']).Family.agg([sum])
animation = df_cat_year.groupby(['Year']).Animation.agg([sum])
history = df_cat_year.groupby(['Year']).History.agg([sum])


# Change the name of the column so it would be more easy to read it 
action = action.rename(columns={'sum': 'Action'})
drama = drama.rename(columns={'sum': 'Drama'})
horror = horror.rename(columns={'sum': 'Horror'})
thriller = thriller.rename(columns={'sum': 'Thriller'})
comedy = comedy.rename(columns={'sum': 'Comedy'})
romance = romance.rename(columns={'sum': 'Romance'})
adventure = adventure.rename(columns={'sum': 'Adventure'})
crime = crime.rename(columns={'sum': 'Crime'})
fantasy = fantasy.rename(columns={'sum': 'Fantasy'})
western = western.rename(columns={'sum': 'Western'})
biography = biography.rename(columns={'sum': 'Biography'})
mystery = mystery.rename(columns={'sum': 'Mystery'})
musical = musical.rename(columns={'sum': 'Musical'})
war = war.rename(columns={'sum': 'War'})
sci_fi = sci_fi.rename(columns={'sum': 'Sci-Fi'})
sport = sport.rename(columns={'sum': 'Sport'})
music = music.rename(columns={'sum': 'Music'})
family = family.rename(columns={'sum': 'Family'})
animation = animation.rename(columns={'sum': 'Animation'})
history = history.rename(columns={'sum': 'History'})


# merging all the dataframes
data = pd.merge(action, drama, left_on='Year', right_on='Year')
data2 = pd.merge(data, horror, left_on='Year', right_on='Year')
data3 = pd.merge(data2, thriller, left_on='Year', right_on='Year')
data4 = pd.merge(data3, comedy, left_on='Year', right_on='Year')
data5 = pd.merge(data4, romance, left_on='Year', right_on='Year')
data6 = pd.merge(data5, adventure, left_on='Year', right_on='Year')
data7 = pd.merge(data6, crime, left_on='Year', right_on='Year')
data8 = pd.merge(data7, fantasy, left_on='Year', right_on='Year')
data9 = pd.merge(data8, western, left_on='Year', right_on='Year')
data10 = pd.merge(data9, biography, left_on='Year', right_on='Year')
data11 = pd.merge(data10, mystery, left_on='Year', right_on='Year')
data12 = pd.merge(data11, musical, left_on='Year', right_on='Year')
data13 = pd.merge(data12, war, left_on='Year', right_on='Year')
data14 = pd.merge(data13, sci_fi, left_on='Year', right_on='Year')
data15 = pd.merge(data14, sport, left_on='Year', right_on='Year')
data16 = pd.merge(data15, music, left_on='Year', right_on='Year')
data17 = pd.merge(data16, family, left_on='Year', right_on='Year')
data18 = pd.merge(data17, animation, left_on='Year', right_on='Year')
data19 = pd.merge(data18, history, left_on='Year', right_on='Year')
data19
```
